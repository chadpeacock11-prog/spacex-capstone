```python
import sqlite3
import pandas as pd

# create in-memory database
conn = sqlite3.connect(':memory:')

# sample data (fake but works)
data = {
    'Launch_Site': ['CCAFS LC-40', 'KSC LC-39A', 'VAFB SLC-4E'],
    'Class': [1, 1, 0]
}

df = pd.DataFrame(data)
df.to_sql('spacex_launch_data', conn, index=False, if_exists='replace')

# run SQL query
query = """
SELECT Launch_Site,
       AVG(Class) * 100 AS Success_Rate
FROM spacex_launch_data
GROUP BY Launch_Site
"""

result = pd.read_sql(query, conn)
print(result)
```
